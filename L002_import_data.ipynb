```python
"""Hur man gör index och slices mot Pandas"""

# Loc är en metod för label-based indexing
# iloc är en metod för integear based indexing (börjar med i - integear)
# [columnnamn] retunerar columnen som en pandas series
```

```python
"""iloc står som sagt för intefear based indexing.
i listan så avser det första talet vilken rad (med start från 0 som vanligt)
där alltså 1 avser rad 2.

Det andra numret avser vilken kolumn där 0 avser första kolumnen.

Det användas för att skriva ut ett visst värde. Retunerar värdet av en enskild cell.
"""

#[x,y] eller [rad,kolumn]
#Funkar även med negativ indexing.

pokemons.iloc[:3, 1]

#Genom loc med en label skriver man såhär exv=



#pokemons
```

```python
#Fredriks dataframe för att träna iloc och loc


column_values = [f"Column{i}" for i in range(1,5)]
index_values = [chr(i) for i in range (65,72)]

dataframe_example = pd.DataFrame([[f"{ind}{col}" for col in range (1,5)] for ind in index_values],
index = index_values,columns= column_values)

dataframe_example
```

```python
#Ger alla rader men kolumn 0.

dataframe_example.iloc[:,0]

# Vusar de tre första raderna för kolumn 1. 
dataframe_example.iloc[:3,0]

dataframe_example.iloc[1, :]

dataframe_example.iloc[:3,0]

dataframe_example.iloc[:2, :2]

dataframe_example.iloc[2:4,0:3]

dataframe_example.loc["A":"B","Column2" : "Column4"]

dataframe_example.iloc[[0,2], [1,3]]

#Första hakparanteserna skriver ut en kolumn och den andra indexerar för att skriva ut en specifikt värde.

dataframe_example["Column1"]["D"]

dataframe_example[["Column1", "Column2"]]
```

```python
#Skapar en kopia av dataframen som inte kommer att ändra originalet.

dataframe_example_copy = dataframe_example.copy()

dataframe_example_copy.iloc[0,:] = "JJ"

#Sätter alla celler till 0. 

#dataframe_example_copy[:] = "Jennifer"

dataframe_example_copy.loc["H"] = ["Hej1", "Hej2", "Hej3", "hej4"]

dataframe_example_copy.loc[:, "Column4"] = 1

dataframe_example_copy

dataframe_example_copy.columns = ["Jennifer", "Filip", "Adrian", "Erik"]

dataframe_example_copy.index = [1,2,3,4,5,6,7,8]

# ,inplace = True ger att originaldataframen uppdateras. 

#.drop för att ta bort någon rad eller kolumn.
```
